## Shows the time spent on running different IVAs

```python
import pickle
import os
import pandas as pd

mat_dim_list = [i for i in range(7, 20)]
iva_type_list = ['G', 'S']

data_summary = []

# Loop through possible combinations of iva_type and mat_dim
for iva_type in iva_type_list:
    for mat_dim in mat_dim_list:
        IVA_file_path = f'iva_out/IVA_{iva_type}_{mat_dim}_output.pkl'
        
        # Check if file exists
        if os.path.exists(IVA_file_path):
            with open(IVA_file_path, 'rb') as file:
                data = pickle.load(file)
                # Append a tuple of (mat_dim, iva_type, time_spent)
                data_summary.append((mat_dim, iva_type, int(data['Time'])))

# Create a DataFrame from the collected data
df = pd.DataFrame(data_summary, columns=['Mat_Dim', 'IVA_Type', 'Time_Spent'])

# Pivot the DataFrame to compare times for 'S' and 'G'
pivot_df = df.pivot(index='Mat_Dim', columns='IVA_Type', values='Time_Spent')

pivot_df
```
