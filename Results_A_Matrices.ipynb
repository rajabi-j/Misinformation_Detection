```python
import pickle

results_path = 'model_results_A.pkl'
with open(results_path, 'rb') as file:
    data = pickle.load(file)

for key, value in data.items():
    print(f"\n{key}:")
    if isinstance(value, dict):  # If the value is a dictionary, print its content
        print(f"Best Model Parameters: {value['Best Model Parameters']}")
        print(f"Train F1 Score: {value['Train']['F1 Score']}")
        print(f"Test F1 Score: {value['Test']['F1 Score']}")
```

```python
import matplotlib.pyplot as plt

# Extracting Test F1 Scores
x_labels = [x for x in range(7, 20)]
g_scores = [data[f"G_{x}"]["Test"]["F1 Score"] for x in x_labels]
s_scores = [data[f"S_{x}"]["Test"]["F1 Score"] for x in x_labels]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x_labels, g_scores, label='IVA_G', marker='o')
plt.plot(x_labels, s_scores, label='IVA_Spice', marker='s')

plt.title('Test F1 Scores by IVA Type')
plt.xlabel('Matrix Size')
plt.ylabel('F1 Score')
plt.xticks(x_labels)
plt.legend()
plt.grid(True)
plt.show()
```
